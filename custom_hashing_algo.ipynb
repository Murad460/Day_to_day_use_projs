```python
def custom_hash(data: str) -> str:
    symbols = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789@#%&*"

    values = [ord(c) for c in data]

    for _ in range(4):
        new_values = []
        for i, v in enumerate(values):
            mixed = (v * 31 + i * 17) % 997
            new_values.append(mixed)
        values = new_values

    hash_out = ""
    for v in values:
        hash_out += symbols[v % len(symbols)]

    return hash_out[:32]
```

```python
print(custom_hash("MuradBlockchain"))
print(custom_hash("hello1"))
print(custom_hash("Hello"))

```

```python
def strong_custom_hash(data: str) -> str:
    symbols = "abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789@#%&*"
    MOD = 2**32

    def rotl(x, r):
        return ((x << r) | (x >> (32 - r))) & 0xFFFFFFFF

    state = [
        0x243F6A88, 0x85A308D3, 0x13198A2E, 0x03707344,
        0xA4093822, 0x299F31D0, 0x082EFA98, 0xEC4E6C89
    ]

    for c in data:
        o = ord(c)
        for i in range(8):
            state[i] ^= o
            state[i] = (state[i] * 2654435761) % MOD
            state[i] = rotl(state[i], (i + o) % 31)

    for _ in range(12):
        for i in range(8):
            a = state[i]
            b = state[(i + 1) % 8]
            c = state[(i + 2) % 8]

            state[i] = (a + (b ^ c)) % MOD
            state[i] = rotl(state[i], 13)
            state[i] = (state[i] * 31) % MOD

    hash_out = ""
    for v in state:
        for _ in range(4):
            hash_out += symbols[v % len(symbols)]
            v //= len(symbols)

    return hash_out[:32]
```
