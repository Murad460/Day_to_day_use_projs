Butterfly Flight Stability Algorithm



FUNCTION antenna():

  ground_angle = sensor.sense().angle()

  butterfly_angle = sensor.sense().angle()



  IF ABS(butterfly_angle - ground_angle) >= 90 THEN

    RETURN -1      

  ELSE

    RETURN stabilize(butterfly_angle, ground_angle)

END FUNCTION





FUNCTION wings_pushing():

  status = antenna()



  IF status != -1 THEN

    wings.move_forward(angle = 180 - 90)

END FUNCTION





FUNCTION wings():

  angle = wings_pushing()



  wings.move_up(angle)

  wings.move_down(angle)

END FUNCTION



```python
import random
import math

# ---------------- Sensor ----------------
class Sensor:
    def sense_angle(self):
        # simulate sensed angle in degrees (0–180)
        return random.uniform(0, 180)


# ---------------- Antenna (Stability Unit) ----------------
class Antenna:
    def __init__(self, sensor):
        self.sensor = sensor

    def check_stability(self):
        ground_angle = self.sensor.sense_angle()
        butterfly_angle = self.sensor.sense_angle()

        diff = abs(butterfly_angle - ground_angle)

        if diff >= 90:
            return -1          # unstable
        else:
            return diff        # stable, return correction value


# ---------------- Wings ----------------
class Wings:
    def move_forward(self, angle):
        print(f"Wings moving forward with angle: {angle:.2f}")

    def move_up(self, angle):
        print(f"Wings moving up with angle: {angle:.2f}")

    def move_down(self, angle):
        print(f"Wings moving down with angle: {angle:.2f}")


# ---------------- Butterfly Controller ----------------
class Butterfly:
    def __init__(self):
        self.sensor = Sensor()
        self.antenna = Antenna(self.sensor)
        self.wings = Wings()

    def wings_pushing(self):
        status = self.antenna.check_stability()

        if status == -1:
            print("Unstable flight  — adjusting wings")
            return None
        else:
            push_angle = 180 - status
            self.wings.move_forward(push_angle)
            return push_angle

    def fly(self):
        angle = self.wings_pushing()

        if angle is not None:
            self.wings.move_up(angle / 2)
            self.wings.move_down(angle / 2)
            print("Stable flight ")
        print("-" * 40)


# ---------------- Run Simulation ----------------
butterfly = Butterfly()

for _ in range(5):
    butterfly.fly()
```
