```python
from enum import Enum


# ---------------- ENUMS ---------------- #

class EmotionIntensity(Enum):
    LOW = 1
    MEDIUM = 2
    HIGH = 3


class MentalState(Enum):
    CALM = "calm"
    BORED = "bored"
    ANXIOUS = "anxious"
    STRESSED = "stressed"
    EXTREME = "extreme"


class EmotionType(Enum):
    NORMAL = "normal"
    FEAR = "fear"
    PAIN = "pain"
    EXTREME_FEAR = "extreme_fear"
    EXTREME_PAIN = "extreme_pain"


# ---------------- CORE ACTION ---------------- #

def imagery_action():
    return "ACTION TRIGGERED BY IMAGERY"


def no_action():
    return "NO ACTION"


# ---------------- STRONG IMAGERY ---------------- #

def strong_imagery(emotion_type, emotion_intensity, goal_of_imagery, why_use_visualization):

    if emotion_intensity == EmotionIntensity.HIGH:
        return imagery_action()

    elif emotion_intensity == EmotionIntensity.MEDIUM:
        if goal_of_imagery and emotion_type != EmotionType.NORMAL:
            return imagery_action()

    else:
        if why_use_visualization:
            return imagery_action()

    return no_action()


# ---------------- MENTAL STATE ---------------- #

def mental_state_check(state):
    return state in (MentalState.CALM, MentalState.BORED)


# ---------------- EDGE CASE ---------------- #

def edge_case_handler(state, emotion_type):
    if state == MentalState.ANXIOUS:
        if emotion_type in (EmotionType.EXTREME_FEAR, EmotionType.EXTREME_PAIN):
            return imagery_action()
    return no_action()


# ---------------- FINAL DECISION ---------------- #

def decide_action(state, emotion_type, emotion_intensity, goal_of_imagery, why_use_visualization):

    if mental_state_check(state):
        return imagery_action()

    action = strong_imagery(
        emotion_type,
        emotion_intensity,
        goal_of_imagery,
        why_use_visualization
    )

    if action == imagery_action():
        return action

    return edge_case_handler(state, emotion_type)


# ---------------- USER INPUT ---------------- #

def get_user_input():

    print("\n--- Motor Imagery Action Prediction ---\n")

    # Mental State
    print("Mental States:", [s.value for s in MentalState])
    state = MentalState(input("Enter mental state: ").lower())

    # Emotion Type
    print("\nEmotion Types:", [e.value for e in EmotionType])
    emotion_type = EmotionType(input("Enter emotion type: ").lower())

    # Emotion Intensity
    print("\nEmotion Intensity Levels:")
    print("1 = LOW, 2 = MEDIUM, 3 = HIGH")
    intensity_input = int(input("Enter emotion intensity (1/2/3): "))
    emotion_intensity = EmotionIntensity(intensity_input)

    # Goal of imagery
    goal_of_imagery = input("\nIs there a clear goal of imagery? (yes/no): ").lower() == "yes"

    # Purpose of visualization
    why_use_visualization = input("Is visualization strongly intended? (yes/no): ").lower() == "yes"

    return (
        state,
        emotion_type,
        emotion_intensity,
        goal_of_imagery,
        why_use_visualization
    )


# ---------------- MAIN ---------------- #

if __name__ == "__main__":
    inputs = get_user_input()

    result = decide_action(*inputs)

    print("\n--- FINAL RESULT ---")
    print(result)
```
