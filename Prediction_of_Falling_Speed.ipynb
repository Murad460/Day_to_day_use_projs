```python
# Constants
GRAVITY = 9.8          # m/s^2
BACKWARD_PUSH = 0.5    # resistance constant


def object_force(surface_area, density):
    """
    Calculates backward push on the object
    """
    push = BACKWARD_PUSH * surface_area * density
    return push


def space_effect(space_density, surface_area, object_density):
    """
    Calculates resistance due to space (air)
    """
    if space_density == "high":
        resistance = object_force(surface_area, object_density)
    else:
        resistance = object_force(surface_area, object_density) * 0.5

    return resistance


def falling_speed(mass, surface_area, object_density, space_density):
    """
    Calculates net falling speed factor
    """
    resistance = space_effect(space_density, surface_area, object_density)
    net_force = (mass * GRAVITY) - resistance

    speed = net_force / mass
    return speed


# Example usage
mass = 2.0              # kg
surface_area = 0.3      # m^2
object_density = 1.2
space_density = "high"

speed = falling_speed(mass, surface_area, object_density, space_density)
print("Predicted falling speed:", speed, "m/s")
```

```python
# Physical constants
GRAVITY = 9.8          # m/s^2
DRAG_COEFF = 0.47      # drag coefficient (sphere-like object)
TIME_STEP = 0.1        # seconds


class Object:
    def __init__(self, mass, surface_area, density):
        self.mass = mass
        self.surface_area = surface_area
        self.density = density


class Space:
    def __init__(self, density):
        self.density = density


def drag_force(space_density, velocity, surface_area):
    """
    Calculates drag (backward push)
    """
    return 0.5 * DRAG_COEFF * space_density * surface_area * velocity**2


def simulate_fall(obj, space, total_time=5):
    velocity = 0.0
    time = 0.0

    while time < total_time:
        gravity_force = obj.mass * GRAVITY
        drag = drag_force(space.density, velocity, obj.surface_area)

        net_force = gravity_force - drag
        acceleration = net_force / obj.mass

        velocity += acceleration * TIME_STEP
        time += TIME_STEP

    return velocity


# Example usage
obj = Object(
    mass=2.0,
    surface_area=0.3,
    density=1.2
)

space = Space(density=1.225)  # air density (kg/m^3)

final_speed = simulate_fall(obj, space)
print("Final falling speed:", round(final_speed, 2), "m/s")
```
